## Indentation and scopes

- Python uses indentation to identify a _scopes_: it means that the code belongs together
- The `tab` button on your keyboard (to the left of Q) inserts one "tabulation", which is a quick way of adding a standard 4 spaces - careful as other editors may behave differently (e.g. enter a `tab` character, called `\t`)

```python
if condition: # the scope below is execute if the condition is true
    execute_code()     # using the underscore `_` to make multi-word names is called _snake case_
    even_more_code()
```
- You can nest scopes
```python
if True:      # truth values (Booleans) in Python are `True` and `False` -- everything is case sensitive
    if second_condition:
        pass  # does nothing -- useful sometimes
    else:     # the scope below is executed if the `second_condition` is false
        execute_this_function()
              # lines starting with `#` are not evaluated ("comments"), their indentation has no impact
```
- You will need "real code" to run your example, not random names as I did above. You know Booleans, so try using just those and see what you get
```python

    True
else:
    False
```
- Hint: when you evaluate a code cell, the output underneath is the last value calculated. If you get no output, it means the last value evaluated to `None`, another keyword which indicates absence of something.
- Note that other languages use semicolons `;` to end a line, while Python just uses a newline. If you need to write multiple statements together though, you _can_ terminate a statement with a semicolon: `print(True); print(True)`

## Conditionals

---

- Conditional: a piece of code (scope) that is executed only if a condition is evaluated to `True`
- The structure is simple
```python
if True:
    False
```
- There is a keyword `if`, a condition to evaluate (here just `True`), a colon `:` terminates the condition, then newline and deeper indentation to indicate a new scope, which is run if the condition is true (here `False`)
- You can also add `else` (without condition) which evaluates its scope if the `if` clause is False
- There is also `elif` which takes a condition: the scope is run if the previous condition is false and the current condition is true
- You can also chain stuff with `and` and `or` and `not`, following classical logics


## Basic data structures

---

- You will be using lists most of the time, but you need to know other types depending on the application
- Ordered collections of elements are called Lists, denoted by `[]` or explicitly `list()`
```python
lst = [1, 2, 1.2, True] # the elements can be of different types (true for all of these structures)
print(lst)
print(sort([3,5,2,4,6])) # if you use homogeneous elements, you can sort a list
```
- Immutable lists are called Tuples, denoted by `()` or explicitly `tuple()`
```python
tpl = (1, 2, 1.2, True) # you cannot add or remove an element to a tuple
print(tpl)
```
- Unordered collections of elements are called Sets, denoted by `{}` or explicitly `set()`
```python
my_set = {1, 2, 1.2, True} # you cannot sort a set
print(my_set) 
```
- Unordered maps, pairing a key to a value, Dictionaries, denoted by `{}` (but elements are `key : value` or explicitly `dict()`
```python
my_map = {'first' : 1, 'second' : 2, 'third' : 3}
print(my_map)
```
- You can assign multiple variables together (more on that later)
```python
a, b = ['first', 'second']
print(b)
```
- You can access the values in _ordered_ types by position using square brackets after the name
```python
print(lst[2])  # the first element is in position zero, so this is the third: beware!
print(tpl[-1]) # negative indices start from the end
print(my_map['second']) # access dictionaries by key: it can be anything!
print(my_set[0])  # what happens here? Why?
```

- What we are doing right now is often called _"scripting"_, which basically means that you hope that your program does not grow enough that you will need to care for it and add structure :)
- There are two loop keywords in Python: 
    - `while` takes a condition like if, and repeats executing its scope until the condition is False
        ```python
        enter_loop = True
        loop_counter = 0
        while enter_loop:
            print('loop count:', loop_counter+1) # careful: in Python we start most things at zero
            loop_counter += 1 # means `loop_counter = loop_counter + 1`
            if loop_counter > 3:
                enter_loop = False
        ```
    - `for` takes a variable name and a sequence, and will bind (`=`) the variable to each of the values in turn
        ```python
        for counter in [1,2,3,4]:
            print('the current counter is', counter)

        print('range() generates a list of integers of given length (starting at zero')
        for i in range(2): print(i) # you can write single-line scopes after the colon symbol `:`
        ```

## Function definition and calling

---

- A function is anything that takes an input, does some elaboration on it, then returns an output
- In Python they are named more broadly "callable", because the built-in `callable()` returns `True` if the argument can be called like a function, regardless of what type it is
- To call a function in Python, you simply add parenthesis after its name (which is just another variable assignment) and write the arguments inside
- In python the correct terminology is to call these values "arguments", but more generally they are also called "parameters", feel free to use the terms as equivalents (I myself typically use the latter)
- The simplest method to define a function is with the built-in keyword `def`, followed by the name, then in parenthesis the argument(s), a colon `:`, and an indented scope with the code (that can use the variable named as arguments)
```python
def add_one(n):
    n += 1 # remember this?
    print(n)   # this function prints the argument and returns `None`
    return n  # this keyword says that `add_one()` returns something that is not None
b = add_one(2)
print("one more:", b+1)
```
